```python
from elasticsearch import Elasticsearch
from elasticsearch import helpers
import pyodbc
import pandas as pd

```

```python
def get_connection(db_config_path):
    with open(db_config_path, encoding='utf8') as db:
        database = ';'.join(db.readlines()[1:]).replace('\n', '').replace(' ', '') + ';'
    return pyodbc.connect('DRIVER={/opt/microsoft/msodbcsql18/lib64/libmsodbcsql-18.3.so.2.1};' + database + 'TrustServerCertificate=yes;')
```

```python
sql = """SELECT PDM.ID_PDM,  MAT.ID_MATERIAL, PDM.NOME_MATERIAL, CAR.CARACTERISTICA, CAR_VAL.CARACTERISTICA_VALOR, SUB.DESCRICAO_SUBCLASSE, CLS.DESCRICAO_CLASSE, GRU.DESCRICAO_GRUPO, DIV.DESCRICAO_DIVISAO, SEC.DESCRICAO_SECAO 
FROM CATALOGO.PDM AS PDM 
LEFT JOIN CATALOGO.CLASSES AS CLS ON PDM.ID_CLASSE = CLS.ID_CLASSE 
LEFT JOIN CATALOGO.SUBCLASSES AS SUB ON PDM.ID_SUBCLASSE = SUB.ID_SUBCLASSE 
left JOIN CATALOGO.GRUPOS AS GRU ON CLS.ID_GRUPO = GRU.ID_GRUPO 
LEFT JOIN CATALOGO.DIVISOES AS DIV ON GRU.ID_DIVISAO = DIV.ID_DIVISAO 
LEFT JOIN CATALOGO.SECOES AS SEC ON DIV.ID_SECAO = SEC.ID_SECAO 
left JOIN CATALOGO.MATERIAIS AS MAT ON PDM.ID_PDM = MAT.ID_PDM 
left JOIN CATALOGO.MATERIAL_CARACTERISTICA AS MAT_CAR ON MAT.ID_MATERIAL = MAT_CAR.ID_MATERIAL 
left join CATALOGO.PDM_CARACTERISTICA AS PDM_CAR ON MAT_CAR.ID_PDM_CARACTERISTICA = PDM_CAR.ID_PDM_CARACTERISTICA 
left JOIN CATALOGO.CARACTERISTICAS AS CAR ON CAR.ID_CARACTERISTICA = PDM_CAR.ID_CARACTERISTICA 
left JOIN CATALOGO.CARACTERISTICAS_VALORES AS CAR_VAL ON PDM_CAR.ID_CARACTERISTICA_VALOR = CAR_VAL.ID_CARACTERISTICA_VALOR
"""
```

```python
sqlMaterial = """SELECT 
            m.ID_MATERIAL, 
            p.TIPO, 
            M.CODIGO_MATERIAL, 
            M.ITEM_SUSPENSO,
            M.MATERIAL_SUSTENTAVEL,
            c.CODIGO_CLASSE,
            c.DESCRICAO_CLASSE,
            s.CODIGO_SUBCLASSE,
            s.DESCRICAO_SUBCLASSE,
            g.CODIGO_GRUPO,
            g.DESCRICAO_GRUPO,
            d.CODIGO_DIVISAO,
            d.DESCRICAO_DIVISAO,
            s2.CODIGO_SECAO,
            s2.DESCRICAO_SECAO,
            ISNULL(STUFF((SELECT ' ' + CONCAT(c.CARACTERISTICA, ' ', CV.CARACTERISTICA_VALOR, cv.SIGLA_UNIDADE_MEDIDA)
                        FROM CATALOGO.MATERIAL_CARACTERISTICA mc
                        JOIN CATALOGO.PDM_CARACTERISTICA pc ON pc.ID_PDM_CARACTERISTICA = mc.ID_PDM_CARACTERISTICA
                        JOIN CATALOGO.CARACTERISTICAS c ON pc.ID_CARACTERISTICA = c.ID_CARACTERISTICA
                        JOIN CATALOGO.CARACTERISTICAS_VALORES cv ON cv.ID_CARACTERISTICA_VALOR = pc.ID_CARACTERISTICA_VALOR
                        WHERE mc.ID_MATERIAL = m.ID_MATERIAL
                        FOR XML PATH('')), 1, 1, ''), '') AS CARACTERISTICAS_BUSCA,
            p.NOME_MATERIAL, 
            CONCAT(p.NOME_MATERIAL, ' ', ISNULL(STUFF((SELECT ' ' + CONCAT(c.CARACTERISTICA, ' ', CV.CARACTERISTICA_VALOR, cv.SIGLA_UNIDADE_MEDIDA)
                                                    FROM CATALOGO.MATERIAL_CARACTERISTICA mc
                                                    JOIN CATALOGO.PDM_CARACTERISTICA pc ON pc.ID_PDM_CARACTERISTICA = mc.ID_PDM_CARACTERISTICA
                                                    JOIN CATALOGO.CARACTERISTICAS c ON pc.ID_CARACTERISTICA = c.ID_CARACTERISTICA
                                                    JOIN CATALOGO.CARACTERISTICAS_VALORES cv ON cv.ID_CARACTERISTICA_VALOR = pc.ID_CARACTERISTICA_VALOR
                                                    WHERE mc.ID_MATERIAL = m.ID_MATERIAL
                                                    FOR XML PATH('')), 1, 1, ''), '')) AS MATERIAL_TEXT
        FROM CATALOGO.MATERIAIS m
        JOIN CATALOGO.PDM p ON m.ID_PDM = p.ID_PDM
        LEFT JOIN CATALOGO.CLASSES c ON p.ID_CLASSE = c.ID_CLASSE
        LEFT JOIN CATALOGO.SUBCLASSES s ON c.ID_CLASSE = s.ID_CLASSE
        LEFT JOIN CATALOGO.GRUPOS g ON g.ID_GRUPO = c.ID_GRUPO
        LEFT JOIN CATALOGO.DIVISOES d ON d.ID_DIVISAO = g.ID_DIVISAO
        LEFT JOIN CATALOGO.SECOES s2 ON S2.ID_SECAO = D.ID_SECAO
        WHERE p.TIPO = 'M'
"""
```

```python
df_materiais = pd.read_sql_query(sqlMaterial, con)
df_materiais
```

```python
es = Elasticsearch(
    "http://localhost:9200",
    request_timeout=999999999,
)

es.info()
```

```python
indices = es.cat.indices(h='index', format='json')
nomes_indices = [indice['index'] for indice in indices]
nomes_indices
```

```python
from sentence_transformers import SentenceTransformer
modelL6V2 = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
embeddings = modelL6V2.encode(df_materiais['MATERIAL_TEXT'].tolist(), show_progress_bar=True)
```

```python
embeddings[5]
```

```python
def doc_generator(index_name, df):
  for idx, row in df.iterrows():
    yield {
        "_index": index_name,
        "_id": idx,
        "_source": {
            "id_material": row['ID_MATERIAL'],
            "tipo": row['TIPO'],
            "codigo_material": row['CODIGO_MATERIAL'],
            "item_suspenso": row['ITEM_SUSPENSO'],
            "material_sustentavel": row['MATERIAL_SUSTENTAVEL'],
            "codigo_classe": row['CODIGO_CLASSE'],
            "descricao_classe": row['DESCRICAO_CLASSE'],
            "codigo_subclasse": row['CODIGO_SUBCLASSE'],
            "descricao_subclasse": row['DESCRICAO_SUBCLASSE'],
            "codigo_grupo": row['CODIGO_GRUPO'],
            "descricao_grupo": row['DESCRICAO_GRUPO'],
            "caracteristica_busca": row['CARACTERISTICAS_BUSCA'],
            "nome_material": row['NOME_MATERIAL'],
            "material_text": row['MATERIAL_TEXT'],
            "vetor_denso": embeddings[idx]
        }
    }

```

```python
INDEX_NAME = "index_s"
documentos_nf_index_generator = doc_generator(INDEX_NAME, df_materiais)
```

```python
es.options(ignore_status=[400,404]).indices.delete(index=INDEX_NAME)

body = {
    "settings": {
        'number_of_shards': 5,
        'number_of_replicas': 1
    },
    'mappings': {
        'properties': {
            'tipo': {'type': 'text' },
            'id_material': {'type': 'long' },
            'codigo_material': {'type': 'keyword' },
            'item_suspenso': {'type': 'text' },
            'material_sustentavel': {'type': 'text' },
            'codigo_classe': {'type': 'keyword' },
            'descricao_classe': {'type': 'text' },
            'codigo_subclasse': {'type': 'keyword' },
            'descricao_subclasse': {'type': 'text' },
            'codigo_grupo': {'type': 'keyword' },
            'descricao_grupo': {'type': 'text' },
            'codigo_divisao': {'type': 'keyword' },
            'descricao_divisao': {'type': 'text' },
            'codigo_secao': {'type': 'keyword' },
            'descricao_secao': {'type': 'text' },
            'caracteristica_busca': {'type': 'text' },
            'nome_material': {'type': 'text' },
            'material_text': {'type': 'text' },
            'vetor_denso': {
                'type': 'dense_vector',
                'dims': 384,
                'index': True,
                'similarity': 'cosine'
            }
        }
    }
}

es.indices.create(index=INDEX_NAME, body=body)
```

```python
indices = es.cat.indices(h='index', format='json')
nomes_indices = [indice['index'] for indice in indices]
nomes_indices
```

```python
import traceback
```

```python
try:
    response = helpers.bulk(es, documentos_nf_index_generator)

    if response[1]:
        errors = response[1]
        for error in errors:
            print("Falha de indexação:", error)
except helpers.BulkIndexError as e:
    for err in e.errors:
        print("Falha de indexação:", err)
    traceback.print_exc()

```

```python
query_embedding = modelL6V2.encode("ssd externo")


query_smt= {
    'field': 'vetor_denso',
    'query_vector': query_embedding.tolist(),
    'k': 10,
    'num_candidates': 50
}

results = es.search(index=INDEX_NAME, knn=query_smt, source=['material_text', 'nome_material'])

results = [
    {
        "_score": hit["_score"],
        "_source": hit["_source"]
    }
    for hit in results["hits"]["hits"]
]

for hit in results:
    print("Score:", hit["_score"])
    print("Source:", hit["_source"])
```

```python
query_embedding = modelL6V2.encode("motherboard")

response = es.search(
    index=INDEX_NAME,
    knn={
        "field": "vetor_denso",
        "query_vector": query_embedding.tolist(),
        "k": 10,
        "num_candidates": 30,
    },
    source=['material_text']
)

response['hits']['hits']
```
