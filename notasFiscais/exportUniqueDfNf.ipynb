```python
df_nf = df_nf.dropna(subset=['DESCRICAO_PRODUTO'])
```

```python
groupByCodigos = df_nf.groupby('CODIGO_BARRA')
contagem_codigos = groupByCodigos.size()
contagem_codigos
sort_groupby_codigos = contagem_codigos.sort_values(ascending=False)
sort_groupby_codigos
```

```python
# Verificar se uma descrição possui diferentes códigos associados
descricao_com_codigos_diferentes = df_nf.groupby('DESCRICAO_PRODUTO')['CODIGO_BARRA'].nunique()
print("Descrições com diferentes códigos associados:")
print(descricao_com_codigos_diferentes[descricao_com_codigos_diferentes > 1])

#Verificar se um código possui diferentes descrições associadas
codigo_com_descricoes_diferentes = df_nf.groupby('CODIGO_BARRA')['DESCRICAO_PRODUTO'].nunique()
print("\nCódigos com diferentes descrições associadas:")
print(codigo_com_descricoes_diferentes[codigo_com_descricoes_diferentes > 1])

```

```python
#considerando como principal o codigo de barra que mais se repete para uma mesma descrição
codigo_barra_principal = df_nf.groupby('DESCRICAO_PRODUTO')['CODIGO_BARRA'].agg(lambda x: x.value_counts().idxmax())
codigo_barra_principal
```

```python
codigos_repetidos = df_nf[df_nf['DESCRICAO_PRODUTO'] == '#1700100#17049010#TRU BRANCO 30']
codigos_repetidos
```

```python
codigo_barra_principal['#1700100#17049010#TRU BRANCO 30']
```

```python
# Selecionar as linhas correspondentes aos códigos de barras que mais se repetem
df_nf_unique = df_nf[df_nf.apply(lambda row: row['CODIGO_BARRA'] == codigo_barra_principal[row['DESCRICAO_PRODUTO']], axis=1)]
df_nf_unique
```
